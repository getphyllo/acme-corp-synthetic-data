# Acme Corp — exploration starter

Quick walkthrough of the six tables. Run top to bottom; everything is local.

```python
import duckdb
con = duckdb.connect('../acme.duckdb', read_only=True)
con.execute('SHOW TABLES').df()
```

## The headline — Crunchwell Louisiana share by quarter

```python
con.execute("""
SELECT SUBSTR(Week,1,4) AS yr,
       ROUND(AVG(Crunchwell_Value_Share)*100, 2) AS la_share_pct
FROM syndicated_weekly
WHERE DMA='LA-DMA' AND Category='RTE Cereal'
GROUP BY 1 ORDER BY 1
""").df()
```

## OSA on Crunchwell Mega in LA — Hurricane Tonya is visible in Nov–Dec 2025

```python
con.execute("""
SELECT SUBSTR(Date,1,7) AS month,
       ROUND(AVG(OSA_Pct), 1) AS osa,
       COUNT(*) AS n
FROM perfect_store
WHERE DMA='LA-DMA' AND SKU IN ('CR002','CR004','CR005')
GROUP BY 1 ORDER BY 1
""").df()
```

## Switching: which LA Acme-loyal HHs are buying competitor brands?

```python
con.execute("""
SELECT Brand, COUNT(*) AS n
FROM household_transactions
WHERE Switching_Flag='Yes' AND DMA='LA-DMA'
GROUP BY 1 ORDER BY n DESC
""").df()
```
